# üåà Agentes Super-Inteligentes: Rob√¥s que Pensam!

Imagine que voc√™ tem um rob√¥ de brinquedo muito esperto! Normalmente, um rob√¥ s√≥ faz o que voc√™ programa, tipo andar em linha reta ou virar √† direita. Mas um agente √© como um rob√¥ com um c√©rebro de verdade! Esse c√©rebro √© um modelo de linguagem grande (LLM), tipo o Gemini. 

Em vez de voc√™ dizer exatamente o que ele tem que fazer, voc√™ s√≥ diz o que voc√™ quer que ele consiga. Por exemplo: 'Rob√¥, descubra qual √© a capital da Fran√ßa'. O rob√¥ (agente) usa o c√©rebro dele para decidir quais ferramentas ele precisa usar para encontrar a resposta. Ele pode usar uma ferramenta de busca na internet (como o DuckDuckGo) para procurar a informa√ß√£o. 

√â como se o rob√¥ fosse para a biblioteca, procurasse o livro certo, encontrasse a p√°gina e lesse a resposta para voc√™! As 'tools' s√£o como as ferramentas na caixa de ferramentas do rob√¥ ou os livros na biblioteca. O agente decide qual ferramenta usar para resolver o problema. Bem legal, n√©?

```
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_core.prompts import ChatPromptTemplate

# Escolha um modelo LLM (tipo o c√©rebro do rob√¥!)
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.0)

# Crie uma ferramenta (tipo uma lupa para procurar coisas!)
search = DuckDuckGoSearchRun()
tools = [search]

# Crie o prompt (diga ao rob√¥ o que voc√™ quer!)
prompt = ChatPromptTemplate.from_messages(
    [("system", "Voc√™ √© um assistente prestativo."),
     ("user", "{input}")]
)

# Crie o agente (o rob√¥ em si!)
agent = create_tool_calling_agent(llm=llm, tools=tools, prompt=prompt)

# Crie o executor do agente (para colocar o rob√¥ para trabalhar!)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Fa√ßa uma pergunta ao rob√¥!
resposta = agent_executor.invoke({"input": "Qual a capital da It√°lia?"})
print(resposta["output"])

#Agora tente perguntar outra coisa! Por exemplo: 'Qual a cor do c√©u?'
```
