# üåà Superpoderes de Rob√¥s: Aprovando as Decis√µes!

Imagine que temos um rob√¥ muito esperto que ajuda a escrever cartas. Esse rob√¥ √© super r√°pido, mas √†s vezes ele precisa da nossa ajuda para ter certeza de que a carta est√° perfeita! 

√â como na escola! O rob√¥ (que chamamos de LangGraph) faz um rascunho da carta, e ent√£o ele para e espera que o professor (voc√™!) d√™ uma olhada. Voc√™ pode dizer: "Sim, est√° √≥timo! Pode enviar!" ou "Hum, espere! Vamos mudar essa parte para ficar ainda melhor!".

Assim, o rob√¥ s√≥ envia a carta depois que voc√™ aprovar. Isso √© muito importante, especialmente se a carta for muito importante! √â como ter um superpoder para garantir que tudo saia perfeito!

No mundo dos computadores, chamamos isso de "Human-in-the-Loop", que significa que o humano (voc√™!) est√° sempre presente para ajudar e tomar as decis√µes importantes.

```python
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 1. Escolha um modelo de IA (como um rob√¥ muito esperto!)
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")

# 2. Crie um modelo para o rob√¥ seguir (como um roteiro!)
prompt = ChatPromptTemplate.from_messages([("system", "Voc√™ √© um escritor de e-mails."),
    ("user", "Escreva um e-mail curto sobre {tema}.")
])

# 3. Diga ao rob√¥ como ele deve responder (apenas com o texto do e-mail!)
output_parser = StrOutputParser()

# 4. Junte tudo!
chain = prompt | llm | output_parser

# 5. Pe√ßa ao rob√¥ para escrever o e-mail (e voc√™ vai aprovar depois!)
tema = "um convite para uma festa de anivers√°rio"
email = chain.invoke({"tema": tema})

print(f"Rascunho do e-mail: {email}")
print("\nProfessor, aprove ou edite o rascunho!")

# Neste c√≥digo, o rob√¥ (LangChain) escreve um rascunho de e-mail e espera a sua aprova√ß√£o! Super legal, n√©?
```
