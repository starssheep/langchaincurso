```
# Configura莽茫o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```
from langchain.chat_models import ChatGoogleGenerativeAI
from langchain.agents import initialize_agent, Tool

# Defina uma ferramenta simples: contar letras
def contar_letras(palavra: str) -> int:
    """Conta quantas letras tem uma palavra."""
    return len(palavra)

# Crie a ferramenta usando LangChain
ferramentas = [
    Tool(name="ContadorDeLetras", func=contar_letras, description="til para contar o n煤mero de letras em uma palavra.")
]

# Escolha o modelo do rob么 (Gemini 2.0 Flash 茅 r谩pido!)
modelo = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0)

# Crie o agente (nosso rob么)
agente = initialize_agent(ferramentas, modelo, agent="zero-shot-react-description", verbose=True)

# D锚 um problema para o rob么!
agente.run("Quantas letras tem a palavra 'elefante'?")
```
