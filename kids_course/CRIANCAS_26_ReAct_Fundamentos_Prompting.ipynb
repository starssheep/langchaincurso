```python
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

# Crie um modelo de linguagem
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.3)

# Crie um prompt
prompt = PromptTemplate.from_template("Voc√™ √© um rob√¥ que precisa pegar uma bola. Qual a cor da bola?")

# Crie uma cadeia LLM
chain = LLMChain(llm=llm, prompt=prompt)

# Execute a cadeia
resposta = chain.run(cor_da_bola="vermelha")

print(resposta)
```
