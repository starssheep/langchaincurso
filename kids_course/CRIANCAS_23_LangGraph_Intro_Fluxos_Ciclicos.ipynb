```python
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")
prompt = ChatPromptTemplate.from_template("Voc√™ deve responder √† pergunta do usu√°rio: {pergunta}")

chain = prompt | llm | StrOutputParser()

pergunta = "Qual a capital do Brasil?"

resposta = chain.invoke({"pergunta": pergunta})

print(f"A resposta √©: {resposta}")
```
