```python
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field

class Brinquedo(BaseModel):
    nome: str = Field(description="Nome do brinquedo")
    cor: str = Field(description="Cor do brinquedo")

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0)

structured_llm = llm.with_structured_output(Brinquedo)

prompt = ChatPromptTemplate.from_messages(
    [("human", "Qual o nome e a cor de um brinquedo?")]
)

chain = prompt | structured_llm

print(chain.invoke({}))
```
