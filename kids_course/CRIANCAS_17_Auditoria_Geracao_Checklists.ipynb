# üåà Checklists M√°gicas para Detetives de Processos!

Imagine que somos detetives, e cada processo (como dar banho no cachorro ou fazer um bolo) √© um mist√©rio! Para resolver o mist√©rio, usamos uma checklist, que nem uma lista de brinquedos que precisamos para brincar. Essa lista nos ajuda a n√£o esquecer nada importante e garantir que tudo saia perfeito! O computador pode nos ajudar a criar essas listas, pensando em tudo que pode dar errado e como evitar. √â como ter um rob√¥ ajudante para ser um super detetive!

```python
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")
prompt = ChatPromptTemplate.from_template("Me ajude a criar uma checklist para verificar se um bolo est√° perfeito. Liste 3 coisas importantes.")
output_parser = StrOutputParser()

chain = prompt | llm | output_parser

print(chain.invoke({}))
```
