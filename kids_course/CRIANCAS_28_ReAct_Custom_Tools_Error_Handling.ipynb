```
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import initialize_agent, AgentType
from langchain.tools import Tool

# Vamos criar uma ferramenta simples (como uma calculadora de somar)
def somar_numeros(a: float, b: float) -> float:
    """Soma dois n√∫meros."""
    return a + b

# Descrever a ferramenta para o rob√¥ entender
ferramenta_soma = Tool(
    name="CalculadoraSoma",
    func=somar_numeros,
    description="√ötil para somar dois n√∫meros. Voc√™ precisa me dar dois n√∫meros."
)

# Escolher o c√©rebro do rob√¥ (um modelo de linguagem)
modelo = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.0)

# Criar o rob√¥ (agente) com a ferramenta
agente = initialize_agent(
    tools=[ferramenta_soma],
    llm=modelo,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,  # Tipo de agente: pensa, age, observa
    verbose=True, # Para ver o que o rob√¥ est√° pensando
)

# Perguntar para o rob√¥ somar dois n√∫meros
resposta = agente.run("Quanto √© 5 + 3?")
print(resposta)
```
