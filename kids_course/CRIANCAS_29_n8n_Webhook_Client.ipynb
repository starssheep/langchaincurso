Imagine que LangChain √© um rob√¥ muito esperto que sabe muitas coisas e pode conversar. n8n √© como uma central telef√¥nica que conecta esse rob√¥ com outros rob√¥s ou servi√ßos, como enviar mensagens para o Slack ou adicionar coisas em uma planilha do Google. 

Um 'webhook' √© como um bilhete que o rob√¥ LangChain envia para a central telef√¥nica n8n. Quando o rob√¥ LangChain precisa fazer algo (como adicionar uma linha em uma planilha), ele escreve um bilhete (o webhook) com todas as informa√ß√µes e envia para n8n. n8n pega esse bilhete e entrega para o rob√¥ certo (Google Sheets, por exemplo) para fazer o trabalho!

Ent√£o, em vez de LangChain ter que saber como falar diretamente com cada rob√¥, ele s√≥ precisa saber como escrever o bilhete (webhook) e enviar para n8n, que cuida do resto!

```python
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains.conversation.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.3)
memory = ConversationBufferMemory()
conversation = ConversationChain(llm=llm, memory=memory, verbose=False)

print(conversation.predict(input="Ol√°! Qual √© o seu nome?"))
print(conversation.predict(input="Qual √© a capital do Brasil?"))
```
