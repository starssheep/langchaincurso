```python
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.0)

prompt_template = ChatPromptTemplate.from_template("Voc√™ √© um auditor. Verifique se a transa√ß√£o {transacao} est√° de acordo com a pol√≠tica: {politica}")

output_parser = StrOutputParser()

chain = prompt_template | llm | output_parser

politica_exemplo = "Gasto m√°ximo com refei√ß√£o: R$50"
transacao_exemplo = "Gasto com jantar: R$60"

resultado = chain.invoke({"politica": politica_exemplo, "transacao": transacao_exemplo})

print(resultado)
```
