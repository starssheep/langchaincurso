```
# Configura莽茫o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import RetrievalQA
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

# Carregue o texto da lei (simulando)
lei_texto = """A Lei diz que todas as empresas devem ser honestas. Se roubar, vai para a cadeia!"""

# Salve o texto em um arquivo (simulando)
with open("lei.txt", "w") as f:
    f.write(lei_texto)

# Carregando o documento
loader = TextLoader("lei.txt")
document = loader.load()

# Dividindo o texto em partes menores
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)
docs = text_splitter.split_documents(document)

# Criando as embeddings (transformando em n煤meros)
embeddings = HuggingFaceEmbeddings()

# Criando o banco de dados de vetores (mem贸ria do rob么)
db = Chroma.from_documents(docs, embeddings)

# Criando o modelo de linguagem (c茅rebro do rob么)
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")

# Criando a chain (como o rob么 pensa)
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=db.as_retriever())

# Fazendo uma pergunta para o rob么
query = "O que acontece se uma empresa roubar?"

# Obtendo a resposta do rob么
print(qa_chain.run(query))

# A resposta do rob么 ser谩 algo como: 'Se roubar, vai para a cadeia!'

# Limpeza (opcional - remove o arquivo)
import os
os.remove("lei.txt")



```
