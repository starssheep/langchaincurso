S√≥ que, ao inv√©s de darmos as frases para ele de qualquer jeito, usamos um 'molde m√°gico' (Prompt Template) para organizar tudo direitinho. √â como um formul√°rio que diz: 'Rob√¥, traduza isso para mim:'.
```python
# Configura√ß√£o Inicial
import os
from dotenv import load_dotenv

# Carregando chaves do arquivo central de scripts
load_dotenv(r'/Users/naubergois/langchaincurso/scripts/.env')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")

prompt_template = ChatPromptTemplate.from_messages([("system", "Voc√™ √© um tradutor profissional."), ("user", "Traduza {texto} para ingl√™s")])

output_parser = StrOutputParser()

chain = prompt_template | llm | output_parser

texto_para_traduzir = "Ol√°, mundo!"

traducao = chain.invoke({"texto": texto_para_traduzir})

print(traducao)
```
